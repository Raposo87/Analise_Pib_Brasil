O que é Renda Per Capita?

A renda per capita é um dos indicadores socioeconômicos que avaliam o grau de desenvolvimento econômico de um determinado lugar. A média é obtida através da divisão do Produto Nacional Bruto (PNB) pelo número total de habitantes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_excel('Dados_Pib.xlsx')
dados.head(10)
```

```python
dados.isnull().sum()
```

```python
# vizualiar grafico em grid

# Cor de fundo
cor_fundo = '#f5f5f5'

# Criando grids:
grid_graficos = sns.FacetGrid(dados, col='Territorialidades', hue='Territorialidades', col_wrap=4)

# Adicionar grafico de linhas para cada grafico:
grid_graficos = grid_graficos.map(plt.bar, 'Ano',	'PIB per capita')

# Deixa apenas os nomes dos estados nos titulos
grid_graficos = grid_graficos.set_titles('{col_name}')

# Subtitulo
grid_graficos = grid_graficos.fig.suptitle('Evolução da renda percapta por estado', fontsize= 18, color='#404040', fontweight=600)

# Ajustando subtitulo
plt.subplots_adjust(top=0.93)
```
